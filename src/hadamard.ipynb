# Function hadamard

## Synopse
Hadamard Transform.

   - **F = iahadamard(f)**
      - Output:
          - **F**: Image.
      - Input:
          - **f**: Image.

## Function code

```python
import numpy as np

def hadamard(f):
    import ia898.src as ia
    f = np.asarray(f).astype(np.float64)
    if len(f.shape) == 1: f = f[:, newaxis]
    (m, n) = f.shape
    A = ia.hadamardmatrix(m)
    if (n == 1):
        F = np.dot(A, f)
    else:
        B = ia.hadamardmatrix(n)
        F = np.dot(np.dot(A, f), np.transpose(B))
    return F
```
